```python
import csv
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

# Created by Abhishek Kapoor
# Book.csv, is a custom generated file 
filename = '~/Book.csv'

# For Ploting, Empty lists for Latitudes and Longitudes
ls, lsf, lo, lof = [], [], [], []
kills = []
deads = []

# Reading the file
with open(filename) as f:
    reader = csv.reader(f)
    for row in reader:
        ls.append(row[0])
        lo.append(row[1])
        kills.append(row[2])
        deads.append(float(row[3]))

#Converting to Float        
lsf = [float(x) for x in ls]
lof = [float(x) for x in lo]

#Function to define the Colors associated with Number of Deaths
def mark_color(death):
    # Yellow for <10, Blue for <=50, Red for >50
    if death <= 10.0:
        return ('yo')
    elif death <= 50.0:
        return ('bo')
    else:
        return ('ro')

kill = np.unique(kills)

a=0
for ll in kill:
    ls=[]
    lo=[]
    de=[]
    for kil, lon, lat, dea in zip(kills, lof, lsf, deads):
        if kil == ll:
            ls.append(lat)
            lo.append(lon)
            de.append(dea)
    plt.figure(figsize=(16,12))
    bmap = Basemap(projection='robin', resolution = 'c', area_thresh = 1000.0, lon_0=0)
    bmap.bluemarble()
    bmap.drawcoastlines()
    bmap.drawcountries()
    bmap.drawmapboundary()
    plt.title("Scatter Plot as per the Number of Deaths\n Yellow<=10, 10<Blue<=50, Red>100")
    a=a+1
    for long, lati, deay in zip(lo, ls, de):
        if deay <= 10.0:
            min_mark = 2.0
        elif deay <= 50.0:
            min_mark = 0.8
        else:
            min_mark = 0.5
        x,y = bmap(long, lati)
        marker_size = deay * min_mark
        if marker_size > 65.0:
            marker_size = 65.0
        marker_color = mark_color(deay)
        bmap.plot(x, y, marker_color, markersize=marker_size)
    bmap.plot(0, 0, 'ko', markersize=0.2, label = ll)
    plt.legend(fontsize=20, loc = 6,markerscale=20, title='Year', numpoints=1)
    plt.savefig("~/test/frame{0}".format(str(a).rjust(3, "0")))
    plt.clf()
```
