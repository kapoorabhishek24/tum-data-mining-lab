```python
# Input the DB to Memory
import pandas as pd
import numpy as np
print("Loading DB...")
dfs_names = pd.read_csv("final_names.csv")
print("DB Read...")
#print(data_file.sheet_names)
#dfs = data_file.parse(data_file.sheet_names[0])
#print("DB Parsed...")
print(type(dfs_names['group']))
```

```python
#Deleted the rows we don't need
dfs_red1 = dfs.loc[dfs['gname'].isin(dfs_names['group'])]
print(len(dfs_red1['gname']))
dfs_red1 = dfs_red1.reset_index(drop=True)
```

```python
#Delete the columns we don't need
chosen_columns = ['extended', 'multiple', 'suicide', 'attacktype1_txt', 'attacktype2_txt',
       'attacktype3_txt', 'targtype1_txt', 'targtype2_txt', 'targtype3_txt', 'weaptype1_txt',
       'weaptype2_txt', 'weaptype3_txt', 'weaptype4_txt', 'nperps', 'nkill', 'nwound', 'ishostkid', 'gname']

dfs_reduced = dfs_red1.copy()
for col in columns:
    if col not in chosen_columns:
        del(dfs_reduced[col])

columns = dfs_red1.columns
reduced_columns = dfs_reduced.columns
print(columns)
print(reduced_columns)
```

```python
#print(dfs['gname'])
print(len(dfs_reduced))
print(len(dfs_red1))
print(len(dfs))
```

```python
categoricalVariableList =['weaptype1_txt', 'attacktype1_txt', 'targtype1_txt']

dfs_categorical = pd.get_dummies(dfs_reduced, columns = categoricalVariableList)
columns = dfs_categorical.columns
#print(columns)

#Extend the one hot encoding for all the second parameters in dfs_categorical
remaining_cat = [('attacktype2_txt','attacktype1_txt'),('attacktype3_txt','attacktype1_txt'),
       ('targtype2_txt','targtype1_txt'), ('targtype3_txt','targtype1_txt'), 
        ('weaptype2_txt','weaptype1_txt'), ('weaptype3_txt','weaptype1_txt'), ('weaptype4_txt','weaptype1_txt')]

#print(dfs_categorical[remaining_cat[0][0]][52019])
#dfs_categorical[remaining_cat[0][0]][52019] = "Grapes"
#print(dfs_categorical[remaining_cat[0][0]][52019])
```

```python
for index, gname in enumerate(dfs_categorical['gname']):
    for tup in remaining_cat:
        if str(dfs_categorical[tup[0]][index]) != '.':
            col = tup[1]+"_"+dfs_categorical[tup[0]][index]
            if dfs_categorical[col][index] == 0:
                dfs_categorical[col][index] = 1     
```

```python
#Delete mroe columns we don't need

dfs_cat_red = dfs_categorical.copy()
for tup in remaining_cat:
    del(dfs_cat_red[tup[0]])

cat_red_columns = dfs_cat_red.columns
print(cat_red_columns)
```

```python
dfs_cat_red.to_csv('terrorism_red_cat.csv')
```

```python
dimensions = []
for cols in cat_red_columns:
    if cols != "gname":
        dimensions.append(cols)

print(dimensions)
group_dict = {}
for name in dfs_names['x']:
    group_dict[(name,-1)] = []
    for index in range(len(dimensions)):
        group_dict[(name,-1)].append(0)
print(group_dict)
```

```python
for index, gname in enumerate(dfs_cat_red['gname']):
    for i in range(len(dimensions)):
        if (dimensions[i] == 'nkill'):
            print(str(dfs_cat_red[dimensions[i]][index]))
        if str(dfs_cat_red[dimensions[i]][index]) == 'nan'  or dfs_cat_red[dimensions[i]][index] < 0 :
            continue
        group_dict[(gname,-1)][i] += dfs_cat_red[dimensions[i]][index] 
```
