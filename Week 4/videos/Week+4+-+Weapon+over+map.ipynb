```python
import csv
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

# Created by Abhishek Kapoor
# Book.csv, is a custom generated file with limited Columns
filename = '/Users/abhishekkapoor/Desktop/week4.csv'

# For Ploting, Empty lists for Latitudes and Longitudes
ls, lsf, lo, lof = [], [], [], []
years = []
regions = []
attacks = []
targets = []
weapons = []


# Reading the file
with open(filename) as f:
    reader = csv.reader(f)
    for row in reader:
        ls.append(row[0])
        lo.append(row[1])
        years.append(row[2])
        regions.append(row[3])
        attacks.append(row[4])
        targets.append(row[5])
        weapons.append(row[6])
        
#Converting to Float        
lsf = [float(x) for x in ls]
lof = [float(x) for x in lo]

for a,z in enumerate(weapons):
    if z == 'Unknown' or z == 'Other':
        weapons[a]= 'Unknown'
    elif z == 'Biological' or z == 'Chemical' or z == 'Radiological' or z == 'Nuclear':
        weapons[a]= 'Biological/Chemical'
    elif z == 'Sabotage Equipment':
        weapons[a]= 'Sabotage Equipment'
    elif z == 'Melee':
        weapons[a]= 'Melee'
    elif z == 'Incendiary':
        weapons[a]= 'Incendiary'
    elif z == 'Firearms':
        weapons[a]= 'Firearms'
    elif z == 'Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)':
        weapons[a]= 'Vehicle'
    elif z == 'Fake Weapons':
        weapons[a]= 'Fake Weapons'
    elif z == 'Explosives/Bombs/Dynamite':
        weapons[a]= 'Explosives/Bombs/Dynamite'

weapon = np.unique(weapons)

a=0
for ll in weapon:
    if ll != 'Unknown':
        ls=[]
        lo=[]
        for wea, lon, lat in zip(weapons, lof, lsf):
            if wea == ll and wea != 'Unknown' and wea != 'Other':
                ls.append(lat)
                lo.append(lon)
        plt.figure(figsize=(16,12))
        bmap = Basemap(projection='robin', resolution = 'c', area_thresh = 1000.0, lon_0=0)
        bmap.bluemarble()
        bmap.drawcoastlines()
        bmap.drawcountries()
        bmap.drawmapboundary()
        plt.title("Type of Weapon used in Terrorist attacks")
        a=a+1
        x, y = bmap(lo, ls)
        bmap.plot(x, y, 'yo', markersize=8, label = ll)
        plt.legend(fontsize=22, loc = 8, markerscale = 3, numpoints=1, title='Weapon Type')
        plt.savefig("/Users/abhishekkapoor/Desktop/Test/frame{0}".format(str(a).rjust(3, "0")), bbox_inches='tight')
        plt.clf()
```
